### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
import matplotlib.pyplot as plt
```

#1. Análisis de facturación



```python
ingreso_t1 = tienda["Precio"].sum()
ingreso_t2 = tienda2["Precio"].sum()
ingreso_t3 = tienda3["Precio"].sum()
ingreso_t4 = tienda4["Precio"].sum()
```

```python
print("El ingreso de la tienda 1:", ingreso_t1)
print("El ingreso de la tienda 2:", ingreso_t2)
print("El ingreso de la tienda 3:", ingreso_t3)
print("El ingreso de la tienda 4:", ingreso_t4)
```

# 2. Ventas por categoría

```python
categoria_t1 = tienda.groupby("Categoría del Producto").size()
categoria_t2 = tienda2.groupby("Categoría del Producto").size()
categoria_t3 = tienda3.groupby("Categoría del Producto").size()
categoria_t4 = tienda4.groupby("Categoría del Producto").size()
print(categoria_t1)
print(categoria_t2)
print(categoria_t3)
print(categoria_t4)
```

```python
print("La categoria más vendida en la Tienda 1:", categoria_t1.idxmax(),"con",categoria_t1.max())
print("La categoria más vendida en la Tienda 2:", categoria_t2.idxmax(),"con",categoria_t2.max())
print("La categoria más vendida en la Tienda 3:", categoria_t3.idxmax(),"con",categoria_t3.max())
print("La categoria más vendida en la Tienda 4:", categoria_t4.idxmax(),"con",categoria_t4.max())
```

# 3. Calificación promedio de la tienda


```python
calif_t1 = tienda["Calificación"].mean().round(2)
calif_t2 = tienda2["Calificación"].mean().round(2)
calif_t3 = tienda3["Calificación"].mean().round(2)
calif_t4 = tienda4["Calificación"].mean().round(2)
```

```python
print("La calificación de la tienda 1:",calif_t1)
print("La calificación de la tienda 2:",calif_t2)
print("La calificación de la tienda 3:",calif_t3)
print("La calificación de la tienda 4:",calif_t4)
```

# 4. Productos más y menos vendidos

```python
ventas_t1 = tienda.groupby("Producto").size().sort_values(ascending=False)
ventas_t2 = tienda2.groupby("Producto").size().sort_values(ascending=False)
ventas_t3 = tienda3.groupby("Producto").size().sort_values(ascending=False)
ventas_t4 = tienda4.groupby("Producto").size().sort_values(ascending=False)
```

```python
print("El producto más vendido en la tienda 1 es:",ventas_t1.idxmax(),"con",ventas_t1.max())
print("El producto más vendido en la tienda 2 es:",ventas_t2.idxmax(),"con",ventas_t2.max())
print("El producto más vendido en la tienda 3 es:",ventas_t3.idxmax(),"con",ventas_t3.max())
print("El producto más vendido en la tienda 4 es:",ventas_t4.idxmax(),"con",ventas_t4.max())
```

```python
print("El producto menos vendido en la tienda 1 es:",ventas_t1.idxmin(),"con",ventas_t1.min())
print("El producto menos vendido en la tienda 2 es:",ventas_t2.idxmin(),"con",ventas_t2.min())
print("El producto menos vendido en la tienda 3 es:",ventas_t3.idxmin(),"con",ventas_t3.min())
print("El producto menos vendido en la tienda 4 es:",ventas_t4.idxmin(),"con",ventas_t4.min())
```

# 5. Envío promedio por tienda

```python
envio_t1 = tienda["Costo de envío"].mean().round(2)
envio_t2 = tienda2["Costo de envío"].mean().round(2)
envio_t3 = tienda3["Costo de envío"].mean().round(2)
envio_t4 = tienda4["Costo de envío"].mean().round(2)
```

```python
print(f"Costo de envío promedio Tienda 1: {envio_t1}")
print(f"Costo de envío promedio Tienda 2: {envio_t2}")
print(f"Costo de envío promedio Tienda 3: {envio_t3}")
print(f"Costo de envío promedio Tienda 4: {envio_t4}")
```

###GRÁFICOS DE LAS TIENDAS

```python
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
ingresos = [ingreso_t1, ingreso_t2, ingreso_t3, ingreso_t4]

plt.figure()
plt.bar(tiendas, ingresos)

plt.title("Ingreso Promedio")
plt.xlabel("Tiendas")
plt.ylabel("Ingreso Promedio")

plt.show()
```

```python
categorias = categoria_t1.index
ventas = categoria_t1.values

plt.figure()
plt.bar(categorias, ventas)

plt.title("Categorias más vendidas en Tienda 1")
plt.xlabel("Categoria")
plt.ylabel("Cantidad de ventas")

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
promedios = [envio_t1, envio_t2, envio_t3, envio_t4]

plt.figure()
plt.bar(tiendas, promedios)

plt.title("Costo de Envio Promedio")
plt.xlabel("Tiendas")
plt.ylabel("Costo de Envío Promedio")

plt.show()
```
